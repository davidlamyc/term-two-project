```python
import pandas as pd
import numpy as np
```

```python
olist_orders_dataset = pd.read_csv('olist_orders_dataset.csv')
olist_order_items_dataset = pd.read_csv('olist_order_items_dataset.csv')
olist_customers_dataset = pd.read_csv('olist_customers_dataset.csv')
```

```python
# get unique customer for each order
olist_orders_with_unique_customer_id = pd.merge(olist_orders_dataset, olist_customers_dataset, on='customer_id', how='left')
olist_orders_with_unique_customer_id
```

```python
olist_order_items_with_unique_customer_id_dataset = pd.merge(olist_order_items_dataset,olist_orders_with_unique_customer_id,on='order_id',how='left')
olist_order_items_with_unique_customer_id_dataset
```

```python
#### break here ####
```

```python
customer_expenditure_dataset = olist_order_items_with_unique_customer_id_dataset.groupby(['customer_unique_id'])['price', 'freight_value'].sum().reset_index()
customer_expenditure_dataset
```

```python
unique_customer_order_count = olist_order_items_with_unique_customer_id_dataset.groupby(['customer_unique_id'])['customer_id'].count().to_frame('order_count')
unique_customer_order_count
```

```python
df = pd.merge(customer_expenditure_dataset,unique_customer_order_count,on='customer_unique_id',how='left')
df
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
y = df['price']
x1 = df['order_count']
```

```python
plt.scatter(x1, y)
plt.xlabel('x', fontsize=20)
plt.ylabel('y', fontsize=20)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
results.summary()
```
