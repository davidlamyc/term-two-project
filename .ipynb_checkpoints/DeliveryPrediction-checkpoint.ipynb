```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
main_df = pd.read_csv('MergedData_7Nov.csv')
reviews_df = pd.read_csv('./raw_data/olist_order_reviews_dataset.csv')
geo_df = pd.read_csv('./raw_data/olist_geolocation_dataset.csv')
```

```python
reviews_df = reviews_df.drop(columns=['review_id','review_comment_title','review_comment_message','review_creation_date','review_answer_timestamp'])
```

```python
geo_df = geo_df.groupby(['geolocation_zip_code_prefix']).mean()
```

```python
main_df = main_df.merge(geo_df, how='left', left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix')
```

```python
main_df = main_df.rename(columns={'geolocation_lat': 'customer_geolocation_lat', 'geolocation_lng': 'customer_geolocation_lng'})
```

```python
main_df = main_df.merge(geo_df, how='left', left_on='seller_zip_code_prefix', right_on='geolocation_zip_code_prefix')
```

```python
main_df = main_df.rename(columns={'geolocation_lat': 'seller_geolocation_lat', 'geolocation_lng': 'seller_geolocation_lng'})
```

```python
main_df = main_df.loc[:, ~main_df.columns.str.contains('^Unnamed')]
```

```python
def haversine_vectorize(lon1, lat1, lon2, lat2):

    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])

    newlon = lon2 - lon1
    newlat = lat2 - lat1

    haver_formula = np.sin(newlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(newlon/2.0)**2

    dist = 2 * np.arcsin(np.sqrt(haver_formula ))
    km = 6367 * dist #6367 for distance in KM for miles use 3958
    return km
```

```python
main_df['customer_seller_distance'] = haversine_vectorize(main_df['seller_geolocation_lng'],main_df['seller_geolocation_lat'],main_df['customer_geolocation_lng'],main_df['customer_geolocation_lat'])
```

```python
main_df = main_df[main_df['review_score'].notna()] # removing reviews scores that are empty
```

```python
y = main_df['delivery_days']
x1 = main_df[['customer_seller_distance','freight_value','state_AC', 'state_AL', 'state_AM', 'state_AP', 'state_BA', 'state_CE',
       'state_DF', 'state_ES', 'state_GO', 'state_MA', 'state_MG', 'state_MS',
       'state_MT', 'state_PA', 'state_PB', 'state_PE', 'state_PI', 'state_PR',
       'state_RJ', 'state_RN', 'state_RO', 'state_RR', 'state_RS', 'state_SC',
       'state_SE', 'state_SP', 'state_TO']]
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
results.summary()
```

```python
y = main_df['delivery_days']
x1 = main_df['price']
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
results.summary()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x1, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train) #training the algorithm
```

```python
#To retrieve the intercept:
print(regressor.intercept_)

#For retrieving the slope:
print(regressor.coef_)
```

```python
y_pred = regressor.predict(X_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
